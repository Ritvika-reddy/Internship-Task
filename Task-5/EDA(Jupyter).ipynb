```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Dataset

```python
df = pd.read_csv("C:/Users/Ritu/OneDrive/ElevateLabs/Task-5/train.csv")
```

Display First Few Rows of Dataset

```python
df.head()
```

Dataset Information and Structure

```python
df.info()
df.describe()
df['Survived'].value_counts()
```

Checking Missing Values

```python
df.dropna()
df.isnull().sum()
```

Univariate Analysis – Age Distribution

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Observation:
Most passengers are between the ages of 20 and 40.
Very few passengers were over 60 years old, indicating that the majority were young adults.

Survival Count Analysis

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

Observation:
The number of passengers who did not survive is higher than the number who survived.
This shows survival rate was less than 50%.

Bivariate Analysis – Survival by Gender

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

Observation:
Female passengers had a much higher survival rate compared to male passengers.
Most males did not survive, while a large number of females survived.

Bivariate Analysis – Survival by Passenger Class

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Class")
plt.show()
```

Observation:
Passengers from the 1st class had the highest survival rate.
3rd class passengers had the lowest survival rate, indicating class influenced survival chances.

Relationship Between Age and Fare

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Age vs Fare")
plt.show()
```

Observation:
Passengers who paid higher fares generally belonged to higher classes.
There is no strong direct relation between age and fare, but high-fare passengers had better survival chances.

Correlation Analysis using Heatmap

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Observation:

Fare and Survived show positive correlation.
Pclass and Survived show negative correlation (lower class → lower survival).
Passenger class strongly influenced survival.

Pairplot for Multivariate Analysis

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']].dropna())
plt.show()
```

Observation:
Pairplot shows relationships between Age, Fare, Class, and Survival.
Higher fare and lower passenger class number (1st class) show better survival probability.

Conclusion:

Exploratory Data Analysis was performed on the Titanic dataset using Python.
The analysis shows that gender, passenger class, and fare played a major role in survival.
Female passengers and first-class passengers had higher survival chances.
Visualization techniques helped identify patterns, trends, and relationships in the dataset.
