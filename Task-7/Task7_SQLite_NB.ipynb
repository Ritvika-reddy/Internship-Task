Install Required Libraries (Pandas and Matplotlib)

```python
import sqlite3
import pandas as pd
```

Create SQLite Database and Sales Table, Insert Sample Data

```python
conn = sqlite3.connect("sales_data.db")
conn.execute("""CREATE TABLE IF NOT EXISTS sales (Product, Quantity, Price)""")
conn.execute("INSERT INTO sales VALUES ('Laptop', 5, 50000)")
conn.execute("INSERT INTO sales VALUES ('Mobile', 10, 20000)")
conn.execute("INSERT INTO sales VALUES ('Tablet', 7, 15000)")
conn.execute("INSERT INTO sales VALUES ('Laptop', 3, 50000)")
conn.execute("INSERT INTO sales VALUES ('Mobile', 5, 20000)")
conn.commit()
print("Database created successfully")
```

Execute SQL Query to Get Total Quantity and Revenue by Product

```python
query = """SELECT product, SUM(quantity) AS Total_Quality, SUM(quantity * price) AS Revenue FROM sales GROUP BY Product"""
df = pd.read_sql_query(query, conn)
print(df)
```

Display Sales Summary Output / Visualize Revenue by Product Using Bar Chart

```python
import matplotlib.pyplot as plt
df.plot(kind='bar', x='Product', y='Revenue', color='skyblue')
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.savefig("sales_chart.png")
plt.show()
print("Chart saved successfully")
```
