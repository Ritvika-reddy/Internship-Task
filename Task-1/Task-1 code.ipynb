Importing Required Library

```python
import pandas as pd
```

Load the Dataset

```python
df=pd.read_csv("C:/Users/Ritu/Downloads/netflix_titles_nov_2019.csv.zip")
df
```

Identifing Missing Values in each Column

```python
df.isnull().sum()
```

Handle Missing Values

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['rating'] = df['rating'].fillna("Not Rated")
df['date_added'] = pd.to_datetime(df['date_added'])
df['date_added'] = df['date_added'].fillna(pd.to_datetime(df['release_year'], format='%Y'))
```

Remove Duplicate Data

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

Checking for the Null Values

```python
df.isnull().sum()
```

Checking for the Duplicate Data

```python
df.duplicated().sum()
```

Standardizing Column Names

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

Removing Extra Spaces and Standardizing Text Formate

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

df['type'] = df['type'].str.lower()
df['country'] = df['country'].str.lower()
df['rating'] = df['rating'].str.lower()
```

Fixing Data Types

```python
df['release_year'] = df['release_year'].astype(int)
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df.to_excel("netflix_movies_and_TVshows.xlsx", index=False)
```
